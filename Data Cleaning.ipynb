The easiest method would be using concatenating code used earlier with addition of renaming columns while joining.
<br>It's time to focus on data inside.
<br>
<br>Having looked at main_df's tail:
<br>#1 There is unnecessary 1230th row which needs to be dropped.
<br>#2 Some countries have unnecessary asterix (*) which could disrupt the data and needs to be removed.

```python

main_df = pd.DataFrame(pd.read_csv(r"./source_files/2015.csv"))
main_df = main_df.rename(columns=correct_names)
main_df['Year'] = '2015'

for i in range(1, len(years)):
    sub_df = pd.read_csv(r"./source_files/"+years[i]+r".csv")
    sub_df = sub_df.rename(columns=correct_names)
    sub_df['Year'] = years[i]
    main_df = pd.concat([main_df,sub_df],axis=0)
    main_df = main_df.reset_index(drop=True)

main_df['Year'] = main_df['Year'].astype(int)
main_df #it allows us to see both head and tail of the dataframe
```

Number of columns is reduced to 29. 
<br>Having looked at main_df.info: 
<br>#1 I only need first 12 columns, with Years included. The rest can be omitted.
<br>#2 I won't need Standard Error column (4th index).
<br>#3 Region and Happiness Rank columns have null values.
* Happiness Rank can be easily calculated based on Happiness Score of the countries (which thankfully there's no null values) 
* Region could be retrieved based on data from other years (none of the countries have moved anywhere and changed their region, I suppose...)

```python
main_df.info()
```

```python

#reducing columns - from 1st to 'Year'
cut_df = main_df.loc[:,:'Year']

#getting rid of Standard Error column and the last row
cut_df = cut_df.drop(columns='Standard Error')
cut_df = cut_df.drop(axis=0, index=1230)

#replacing null value by country Emirates mean of Corruption Perception column
emirates = cut_df[cut_df['Country'].str.contains('Emirates')]
cut_df['Corruption Perception'] = cut_df['Corruption Perception'].fillna(emirates['Corruption Perception'].mean())

#getting rid of asterix from countries' names
cut_df['Country'] = cut_df['Country'].str.replace('*','', regex=True)

cut_df.tail()
```

Let's focus on null values on Region and Happiness Rank.
<br>We clearly see that 2017-2019 and 2022 are missing data on Region and 2020-2021 are missing data on Happiness Rank:

```python
counted_by_year = cut_df.groupby('Year').count()
counted_by_year
```

```python
#checking indexing of 2020-2021 years - 782:1083
missing_rank = cut_df[cut_df.Year.isin([2020,2021])]
missing_rank
```

```python
#calculating Happiness Rank by grouping Happiness Score in descending order 
missing_rank['Happiness Rank'] = missing_rank.groupby('Year')['Happiness Score'].rank('dense',ascending=False)
missing_rank
cut_df.loc[782:1083,'Happiness Rank'] = missing_rank['Happiness Rank']
cut_df[cut_df.Year.isin([2020,2021])]
```

Happiness Rank is fixed. Let's dive in the Region column to see if the naming is consistent among the years and if not, to pick one year which we'll use as a base for all rows:

```python
#checking if the Region values are the same of different among years
region_naming = cut_df.groupby('Year')['Region'].unique()
pd.set_option('display.max_colwidth', 300)
region_naming
```

2015-2016 and 2020-2021 regions are named differenly. 
<br>I chose the first option to be used in my final dataset:

```python
#checking indexing of 2015 - 0:157
cut_df[cut_df.Year.isin([2015])]
```

```python
c = cut_df.loc[0:157].Country.tolist()
r = cut_df.loc[0:157].Region.tolist()
region = {c[i] : r[i] for i in range(len(c))} #2015 'country : region' dictionary to be used in all years

cut_df['Region'] = cut_df.Country.map(region)
cut_df
```

There are still countries with missing Region. I checked each of below whether that was caused by different naming or missing region in 2015 year:

```python
cut_df[cut_df['Region'].isna()].Country.unique()
```

```python
#updating names on countries that had different naming throughout years 
country_correction = {
    'Somaliland Region': 'Somaliland region',
    'Taiwan Province of China': 'Taiwan',
    'Hong Kong S.A.R., China': 'Hong Kong',
    'Trinidad & Tobago': 'Trinidad and Tobago',
    'Northern Cyprus': 'North Cyprus',
    'North Macedonia': 'Macedonia',
    'Hong Kong S.A.R. of China': 'Hong Kong',
    'Czechia': 'Czech Republic'
}
cut_df['Country'] = cut_df.Country.replace(country_correction)

#updating missing region
region_correction = {
    'Puerto Rico': 'Latin America and Caribbean',
    'Belize': 'Latin America and Caribbean',
    'Somalia': 'Sub-Saharan Africa',
    'Namibia': 'Sub-Saharan Africa',
    'South Sudan': 'Sub-Saharan Africa',
    'Gambia': 'Sub-Saharan Africa',
    'Maldives': 'Southern Asia',
    'Congo': 'Sub-Saharan Africa',
    'Eswatini. Kingdom of': 'Sub-Saharan Africa'
}
region.update(region_correction)
cut_df['Region'] = cut_df.Country.map(region)
```

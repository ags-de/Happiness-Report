```python

main_df = pd.DataFrame(pd.read_csv(r"C:/Users/Agata/Documents/GitHub/Happiness-Report/source_files/2015.csv"))
main_df = main_df.rename(columns=correct_names)
main_df['Year'] = '2015'

for i in range(1, len(years)):
    sub_df = pd.read_csv(r"C:/Users/Agata/Documents/GitHub/Happiness-Report/source_files/"+years[i]+r".csv")
    sub_df = sub_df.rename(columns=correct_names)
    sub_df['Year'] = years[i]
    main_df = pd.concat([main_df,sub_df],axis=0)
    main_df = main_df.reset_index(drop=True)
    sub_df.to_csv(r"C:/Users/Agata/Documents/"+years[i]+r".csv", index=True)

main_df['Year'] = main_df['Year'].astype(int)
main_df #it allows us to see both head and tail of the dataframe
```

```python
cut_df = main_df.loc[:,:'Year']
cut_df = cut_df.drop(columns='Standard Error')
cut_df = cut_df.drop(axis=0, index=1230)
cut_df['Country'] = cut_df['Country'].str.replace('*','')
cut_df.loc[cut_df.Year.isin([2020,2021])]
```

```python
rank = cut_df[cut_df.Year.isin([2020,2021])]
rank['Happiness Rank'] = rank.groupby('Year')['Happiness Score'].rank('dense',ascending=False)
#rank.to_csv(r'C:/Users/Agata/Desktop/test_rank.csv', index=True)
cut_df.loc[782:1083,'Happiness Rank'] = rank['Happiness Rank']
cut_df[cut_df.Year.isin([2020,2021])]
```
